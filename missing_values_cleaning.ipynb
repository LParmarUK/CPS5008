```python
import pandas as pd

df = pd.read_csv("uk_employee_records.csv")

print("=== First 5 rows ===")
display(df.head())

print("\n=== Dataset Structure (info) ====")
df.info()

print("\n=== Numerical Summary (describe) ===")
df.describe()

print("\n=== Missing values per column (before) ===")
df.isna().sum()

# Define Columns
numerical_cols = ["Annual Salary (£)", "Years of Experience"]
categorical_cols = ["Gender", "Department", "Employment Type"]

# Impute missing values
# - Numerical: median
# - Categorical: mean
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\n=== Missing values per column (after)")
display(df.isna().sum())

# Valiate distributions remain sensible
print("\n=== Numerical summary after imputation ===")

print("\n=== Categorical distributions after imputation ===")
for col in categorical_cols:
    print(f"\n{col} value counts")
    display(df[col].value_counts())

# Remove duplicates
print("\n=== Duplicate rows before removal ===")
dupes_before = df.duplicated().sum()
print(dupes_before)

df = df.drop_duplicates()

print("\n=== Duplicate rows after removal ===")
dupes_after = df.duplicated().sum()
print(dupes_after)

# Save cleaned dataset
df.to_csv("cleaned_uk_employee_records.csv", index=False)
print("\nSaved: cleaned_uk_employee_records.csv")
```
